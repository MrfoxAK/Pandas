# Pivot Table

```python
import pandas as pd
df = pd.read_csv('student_result1.csv')
df
```

```python
df.pivot_table(index='Section')
```

```python
df
```

```python
df.pivot_table(index='Section',columns='Class')
```

```python
df.pivot_table(index='Section',columns='Class',aggfunc='count')
```

```python
df.pivot_table(index='Section',columns='Class',aggfunc='count',fill_value=0)
```

```python
df.pivot_table(index='Section',columns='Class',aggfunc='count',fill_value=0,margins=True)
```
