# fillna method

```python
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
df
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({"train no": 5,"Name": 4})
```

```python
# for the prev value use ffill & pad
df.fillna(method='ffill')
```

```python
df.fillna(method='pad')
```

```python
# for next value
df.fillna(method='bfill')
```

```python
# axis
df.fillna(method="ffill",axis=0) #for row
```

```python
# axis
df.fillna(method="ffill",axis=1) #for col
```

```python
# axis
df.fillna(method="bfill",axis=0) #for row
```

```python
df.fillna(0,limit=1)
```

```python
df.fillna(method="ffill",limit=1)
```

```python
df.fillna(4,inplace= True)
df
```
