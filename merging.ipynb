# pandas merging function

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({"ID":[1,2,4,3,46],"Class":[9,10,11,12,44]})
df1
```

```python
df2 = pd.DataFrame({"ID":[1,2,4,3,14],"Name":['AK','BK','CK','DK','EK']})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='ID')
```

```python
pd.merge(df2,df1)
```

```python
pd.merge(df1,df2,on='ID',how='left') #left dataframe df1
```

```python
pd.merge(df1,df2,on='ID',how='right') #left dataframe df2
```

```python
pd.merge(df1,df2,on='ID',how='outer')
```

```python
pd.merge(df1,df2,on='ID',how='outer',indicator=True)
```

```python
df2 = pd.DataFrame({"ID":[11,12,14,31,114],"Name":['AK','BK','CK','DK','EK']})
df2
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True,suffixes=('-df1','-df2'))
```
